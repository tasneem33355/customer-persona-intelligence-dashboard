```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

df = pd.read_csv("bank-additional-full.csv", sep=';')

print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum(), (df == "unknown").sum().sort_values(ascending=False)
```

```python
df = df.drop_duplicates().reset_index(drop=True)

categorical_cols = df.select_dtypes(include="object").columns

df[categorical_cols] = df[categorical_cols].replace("unknown", "Unknown")

print(df.shape)
```

```python
y = df["y"]
X = df.drop(columns=["y"])

X.shape
```

```python
num_cols = X.select_dtypes(include=["int64", "float64"]).columns

X[num_cols].describe()
```

```python
X[num_cols].hist(bins=30)
plt.tight_layout()
plt.show()
```

```python
X["duration_log"] = np.log1p(X["duration"])
X["campaign_log"] = np.log1p(X["campaign"])

X = X.drop(columns=["duration", "campaign"])

X.shape
```

```python
cat_cols = X.select_dtypes(include="object").columns

pd.DataFrame({
    "unique_values": X[cat_cols].nunique(),
    "sample_values": X[cat_cols].apply(lambda x: x.unique()[:5])
}).sort_values(by="unique_values", ascending=False)
```

```python
X["campaign_intensity"] = X["previous"] / (X["campaign_log"] + 1)
X["contact_effort"] = X["campaign_log"] * X["duration_log"]
X["economic_pressure"] = (
    X["emp.var.rate"] +
    X["euribor3m"] +
    X["nr.employed"] / 1000
)

X[["campaign_intensity", "contact_effort", "economic_pressure"]].describe()
```

```python
cat_cols = X.select_dtypes(include="object").columns
num_cols = X.select_dtypes(exclude="object").columns

X_encoded = pd.get_dummies(X, columns=cat_cols, drop_first=True)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

X_scaled.shape
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia = []
silhouette = []

K_range = range(2, 9)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(X_scaled, labels))

list(zip(K_range, inertia, silhouette))
```

```python
kmeans = KMeans(n_clusters=2, random_state=RANDOM_STATE, n_init=10)
kmeans_labels = kmeans.fit_predict(X_scaled)

np.unique(kmeans_labels, return_counts=True)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam

input_dim = X_scaled.shape[1]
encoding_dim = 10

input_layer = Input(shape=(input_dim,))
encoded = Dense(32, activation='relu')(input_layer)
encoded = Dense(encoding_dim, activation='relu')(encoded)
decoded = Dense(32, activation='relu')(encoded)
decoded = Dense(input_dim, activation='linear')(decoded)

autoencoder = Model(inputs=input_layer, outputs=decoded)
encoder = Model(inputs=input_layer, outputs=encoded)

autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

autoencoder.summary()
```

```python
history = autoencoder.fit(
    X_scaled, X_scaled,
    epochs=50,
    batch_size=256,
    shuffle=True,
    validation_split=0.1,
    verbose=1
)
```

```python
X_embedded = encoder.predict(X_scaled)
X_embedded.shape
```

```python
inertia_emb = []
silhouette_emb = []

for k in range(2, 9):
    kmeans = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=10)
    labels = kmeans.fit_predict(X_embedded)
    inertia_emb.append(kmeans.inertia_)
    silhouette_emb.append(silhouette_score(X_embedded, labels))

list(zip(range(2, 9), inertia_emb, silhouette_emb))
```

```python
kmeans_deep = KMeans(n_clusters=2, random_state=RANDOM_STATE, n_init=10)
deep_labels = kmeans_deep.fit_predict(X_embedded)

np.unique(deep_labels, return_counts=True)
```

```python
from sklearn.mixture import GaussianMixture

gmm_silhouette = []
gmm_bic = []
gmm_aic = []

for k in range(2, 9):
    gmm = GaussianMixture(
        n_components=k,
        covariance_type="full",
        random_state=RANDOM_STATE
    )
    labels = gmm.fit_predict(X_embedded)
    gmm_silhouette.append(silhouette_score(X_embedded, labels))
    gmm_bic.append(gmm.bic(X_embedded))
    gmm_aic.append(gmm.aic(X_embedded))

list(zip(range(2, 9), gmm_silhouette, gmm_bic, gmm_aic))
```

```python
gmm_final = GaussianMixture(
    n_components=5,
    covariance_type="full",
    random_state=RANDOM_STATE
)

gmm_labels = gmm_final.fit_predict(X_embedded)

np.unique(gmm_labels, return_counts=True)
```

```python
business_features = [
    "age",
    "duration_log",
    "previous",
    "campaign_log",
    "contact_effort",
    "campaign_intensity",
    "job",
    "education",
    "housing",
    "loan"
]

X_business = X[business_features].copy()
X_business.shape
```

```python
X_business["engagement_score"] = (
    X_business["duration_log"] +
    X_business["contact_effort"]
)

X_business["persistence_score"] = (
    X_business["campaign_log"] +
    X_business["previous"]
)

X_business["financial_exposure"] = (
    (X_business["housing"] == "yes").astype(int) +
    (X_business["loan"] == "yes").astype(int)
)

X_business[["engagement_score", "persistence_score", "financial_exposure"]].describe()
```

```python
def assign_persona(row):
    if row.engagement_score > 12.5 and row.persistence_score > 1.6:
        return "Highly Engaged Loyalist"
    
    if row.engagement_score > 12.5 and row.financial_exposure == 0:
        return "Curious Safe Explorer"
    
    if row.engagement_score < 8.8 and row.persistence_score < 0.69:
        return "Cold & Uninterested"
    
    if row.financial_exposure == 2 and row.persistence_score > 1.1:
        return "Financially Stressed Repeater"
    
    return "Moderate Potential"
```

```python
X_business["persona"] = X_business.apply(assign_persona, axis=1)
X_business["persona"].value_counts(normalize=True) * 100
```
